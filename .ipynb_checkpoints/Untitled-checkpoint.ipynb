```python
# Libraries for OCR
import pytesseract
from pdf2image import convert_from_path
from PIL import Image
```

```python
# sample pdf for testing
pdf_path = "data/60-1234-00-PP220-0008.pdf"
```

```python
# pytesseract only works on image not on pdf, therefore, converting pdf to image of size 600x600
page = convert_from_path(pdf_path, 600)
```

```python
# page[0] stores image, page[1] stores meta data, extracting only image
page[0]
```

```python
# saving image as jpg for later use
page[0].save('img.jpg')
```

```python
# calling pre-installed pytesseract.exe application from installation directory for usage
pytesseract.pytesseract.tesseract_cmd = "C:/Program Files/Tesseract-OCR/tesseract.exe"
```

```python
# identifying text from image
text = pytesseract.image_to_string(page[0])
```

```python
# text extracted from image
text
```

```python
# replacing special characters
updated = text.replace('\n', ' ')
updated = updated.replace(b'\xe2\x80\x94'.decode('utf-8'),'')
```

```python
# after replacing special characters
updated
```

```python
# splitting text in word array
t = updated.split(' ')
t
```

```python
# matching drawing name in wolders with words in arrays
drawing_list = []
for i in t:
    for j in references:
        if i == j:
            drawing_list.append(i)
```

```python
def create_path(name_array, root_dir):
    path_arr = []
    for names in name_array:
        single_path = root_dir + names + '.pdf'
        path_arr.append(single_path)
    
    return path_arr
    
```

```python
doc = fitz.open("data/60-1234-00-PP220-0008.pdf")           # or new: fitz.open(), followed by insertPage()
page = doc[0]                         # choose some page
rect = fitz.Rect(50, 100, 300, 400)   # rectangle (left, top, right, bottom) in pixels

text = """This text will only appear in the rectangle. Depending on width, new lines are generated as required.\n<- This forced line break will also appear.\tNow a very long word: abcdefghijklmnopqrstuvwxyzABCDEFGHIJKLMNOPQRSTUVWXYZ.\nIt will be broken into pieces."""

rc = page.insertTextbox(rect, text, fontsize = 12, # choose fontsize (float)
                   fontname = "Times-Roman",       # a PDF standard font
                   fontfile = None,                # could be a file on your system
                   align = 0)                      # 0 = left, 1 = center, 2 = right

print("unused rectangle height: %g" % rc)          # just demo (should display "44.2")

doc.saveIncr()
```

```python
import PyPDF2
```

```python
watermark_pdf_path = "data/60-1234-00-PP220-0008.pdf"
pdf_path = "data/60-1234-00-CC100-0007.pdf"

map_pdf = open(pdf_path, 'rb')
watermark_pdf = open(watermark_pdf_path, 'rb')

pdfReader = PyPDF2.PdfFileReader(map_pdf)
pdfWatermarkReader = PyPDF2.PdfFileReader(watermark_pdf)
pdfWriter = PyPDF2.PdfFileWriter()

map_pdf_first_page = pdfReader.getPage(0)
map_pdf_first_page.mergePage(pdfWatermarkReader.getPage(0))
pdfWriter.addPage(map_pdf_first_page)

resultPdfFile = open(r"C:\Users\44745\Desktop\Hack_11\data\Complete.pdf", 'wb')

pdfWriter.write(resultPdfFile)
map_pdf.close()
watermark_pdf.close()
resultPdfFile.close()
```
